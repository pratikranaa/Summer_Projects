```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.forecasting.theta import ThetaModel
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.model_selection import train_test_split
import warnings
import matplotlib.pyplot as plt
```

```python
warnings.filterwarnings("ignore")
```

```python
# Load the data
file_path = '100_SKUS.csv'
data = pd.read_csv(file_path)
data_long = pd.melt(data, id_vars=['material'], var_name='date', value_name='sales')
data_long['date'] = pd.to_datetime(data_long['date'], format='%m/%d/%Y')
data_long = data_long.sort_values(by=['material', 'date'])
```

```python
data_long
```

```python
# Prepare for model evaluation
def create_features(df, label=None):
    df['month'] = df.index.month
    df['year'] = df.index.year
    X = df[['month', 'year']]
    if label:
        y = df[label]
        return X, y
    return X

def evaluate_forecasts(actual, predicted):
    mape = mean_absolute_percentage_error(actual, predicted) * 100
    accuracy = 100 - mape
    return accuracy
```

```python
results = []

# Function to train and test models
def train_and_test_models(train, test, material, month):
    models = {
        'ARIMA': ARIMA(train, order=(2, 1, 2)),  # Replace with appropriate parameters
        'SARIMA': SARIMAX(train, order=(2, 1, 2), seasonal_order=(1, 1, 1, 12)),  # Replace with appropriate parameters
        'Theta': ThetaModel(train),
        'XGBoost': XGBRegressor(objective='reg:squarederror'),
        'CatBoost': CatBoostRegressor(silent=True),
        'Linear Regression': LinearRegression()
    }

    # Train and forecast
    forecasts = {}
    for name, model in models.items():
        if name in ['ARIMA', 'SARIMA', 'Theta']:
            model_fit = model.fit()
            forecast = model_fit.forecast(steps=len(test))
        else:
            X_train, y_train = create_features(train.to_frame(), label='sales')
            X_test, y_test = create_features(test.to_frame(), label='sales')
            model.fit(X_train, y_train)
            forecast = model.predict(X_test)
        
        forecast[forecast < 0] = 0  # Set negative forecasts to zero
        forecasts[name] = forecast

    # Evaluate models
    accuracies = {name: evaluate_forecasts(test, forecast) for name, forecast in forecasts.items()}

    # Save results for all models
    for name, accuracy in accuracies.items():
        results.append({
            'material': material,
            'month': month,
            'model': name,
            'accuracy': accuracy,
            'actual': test.values,
            'predicted': forecasts[name]
        })

# Train and test with rolling windows
materials = data_long['material'].unique()
for material in materials:
    sku_data = data_long[data_long['material'] == material].set_index('date')['sales']
    dates = pd.date_range(start='2023-07-01', end='2023-12-01', freq='MS')

    for date in dates:
        train = sku_data[:date - pd.DateOffset(months=1)]
        test = sku_data[date:date + pd.DateOffset(months=1) - pd.DateOffset(days=1)]
        if len(test) > 0:
            train_and_test_models(train, test, material, date.strftime('%Y-%m'))

# Create DataFrame for results
results_df = pd.DataFrame(results)
```

```python
# Compute average accuracies for each model for each SKU
average_accuracies_df = results_df.groupby(['material', 'model'])['accuracy'].mean().reset_index()

# Identify the best model for each SKU
best_models_df = average_accuracies_df.loc[average_accuracies_df.groupby('material')['accuracy'].idxmax()]

# Print or return the aggregated results
print("Average Accuracies:")
print(average_accuracies_df)
print("\nBest Models:")
print(best_models_df[['material', 'model', 'accuracy']])


results_df.to_csv("result.csv")
average_accuracies_df.to_csv("average_accuracies.csv")# Create DataFrame for results
best_models_df.to_csv("best_models.csv")
```

```python
best_model_forecasts = []

for index, row in best_models_df.iterrows():
    material = row['material']
    best_model_name = row['model']

    sku_data = data_long[data_long['material'] == material].set_index('date')['sales']
    train = sku_data[:'2023-12-31']
    
    if best_model_name == 'ARIMA':
        model = ARIMA(train, order=(2, 1, 2))  # Replace with actual best parameters
    elif best_model_name == 'SARIMA':
        model = SARIMAX(train, order=(2, 1, 2), seasonal_order=(1, 1, 1, 12))  # Replace with actual best parameters
    # elif best_model_name == 'ETS':
    #     model = ExponentialSmoothing(train, seasonal='add', seasonal_periods=12)
    elif best_model_name == 'Theta':
        model = ThetaModel(train)
    elif best_model_name == 'XGBoost':
        X_train, y_train = create_features(train.to_frame(), label='sales')
        model = XGBRegressor(objective='reg:squarederror')
        model.fit(X_train, y_train)
    elif best_model_name == 'CatBoost':
        X_train, y_train = create_features(train.to_frame(), label='sales')
        model = CatBoostRegressor(silent=True)
        model.fit(X_train, y_train)
    elif best_model_name == 'Linear Regression':
        X_train, y_train = create_features(train.to_frame(), label='sales')
        model = LinearRegression()
        model.fit(X_train, y_train)

    # Forecasting for February 2024
    if best_model_name in ['ARIMA', 'SARIMA', 'ETS', 'Theta']:
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=2)
        print(forecast)
        print(best_model_name)
        forecast_value = forecast.iloc[-1]  # February 2024
    else:
        future_dates = pd.date_range(start='2023-12-31', periods=2, freq='MS')
        X_future = create_features(pd.DataFrame(index=future_dates))
        forecast = model.predict(X_future)
        print(forecast)
        print(best_model_name)
        forecast_value = forecast[-1]  # February 2024

    best_model_forecasts.append({
        'material': material,
        'best_model': best_model_name,
        'forecast_value': max(0, forecast_value)  # Ensure no negative forecast
    })

# Create DataFrame for best model forecasts
best_model_forecasts_df = pd.DataFrame(best_model_forecasts)
```

```python
import pandas as pd
from darts import TimeSeries
from darts.models import Croston

results = []


# Function to train and test models
def train_and_test_models(train, test, material, month):
    # Convert train and test data to TimeSeries
    train_series = TimeSeries.from_series(train, fill_missing_dates=True, freq='MS')
    test_series = TimeSeries.from_series(test, fill_missing_dates=True, freq='MS')
    
    # Train the model
    model = Croston(version="optimized")
    model.fit(train_series)
    
    # Make a forecast
    forecast = model.predict(1).values().flatten()
    
    # Set negative forecasts to zero
    forecast[forecast < 0] = 0
    
    # Evaluate the forecast
    accuracy = evaluate_forecasts(test_series.values(), forecast)
    
    # Save results
    results.append({
        'material': material,
        'month': month,
        'model': 'Croston',
        'accuracy': accuracy,
        'actual': test_series.values().flatten(),
        'predicted': forecast
    })

# Train and test with rolling windows
materials = data_long['material'].unique()
for material in materials:
    sku_data = data_long[data_long['material'] == material].set_index('date')['sales']
    dates = pd.date_range(start='2023-07-01', end='2023-12-01', freq='MS')

    for date in dates:
        train = sku_data[:date - pd.DateOffset(months=1)]
        test = sku_data[date:date + pd.DateOffset(months=1) - pd.DateOffset(days=1)]
        
        if len(test) > 0:
            train_and_test_models(train, test, material, date.strftime('%Y-%m'))

# Create DataFrame for results
results_croston_df = pd.DataFrame(results)
results_croston_df
```
