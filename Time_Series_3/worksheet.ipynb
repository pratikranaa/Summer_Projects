```python
import pandas as pd

# Load the data
file_path = '100_SKUS.csv'
data = pd.read_csv(file_path)

data_info = data.info()

data_info, data
```

```python
import matplotlib.pyplot as plt

# Convert data from wide to long format
data_long = pd.melt(data, id_vars=['material'], var_name='date', value_name='sales')

# Convert the 'date' column to datetime format
data_long['date'] = pd.to_datetime(data_long['date'], format='%m/%d/%Y')

# Sort the data by 'material' and 'date'
data_long = data_long.sort_values(by=['material', 'date'])

# Check for missing values
missing_values = data_long.isnull().sum()

# Plot sales data for a few SKUs to visualize trends and seasonality
sample_skus = data_long['material'].unique()[:5]
plt.figure(figsize=(15, 10))
for sku in sample_skus:
    plt.plot(data_long[data_long['material'] == sku]['date'],
             data_long[data_long['material'] == sku]['sales'], label=sku)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Data for Sample SKUs')
plt.legend()
plt.show()

missing_values, data_long.head()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Select a sample SKU for ACF and PACF analysis
sample_sku = 'sku1009'
sample_data = data_long[data_long['material'] == sample_sku].set_index('date')['sales']

# Plot ACF and PACF
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(sample_data, ax=plt.gca(), lags=20)
plt.title(f'ACF for {sample_sku}')

plt.subplot(122)
plot_pacf(sample_data, ax=plt.gca(), lags=20)
plt.title(f'PACF for {sample_sku}')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Augmented Dickey-Fuller test for stationarity
adf_test = adfuller(sample_data)
print('ADF Statistic:', adf_test[0])
print('p-value:', adf_test[1])

# If the series is not stationary, apply differencing
if adf_test[1] > 0.05:
    sample_data_diff = sample_data.diff().dropna()
else:
    sample_data_diff = sample_data
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA


# Fit ARIMA model
arima_model = ARIMA(sample_data, order=(3, 1, 2))  # Replace p, d, q with determined values
arima_fit = arima_model.fit()
print(arima_fit.summary())

sarima_model = SARIMAX(sample_data, order=(3, 1, 2), seasonal_order=(3, 1, 2, 12))  # Replace P, D, Q, s with determined values
sarima_fit = sarima_model.fit()
print(sarima_fit.summary())

# Forecasting
forecast_steps = 36  # Number of steps to forecast
arima_forecast = arima_fit.get_forecast(steps=forecast_steps)
sarima_forecast = sarima_fit.get_forecast(steps=forecast_steps)

# Post-process forecasts to avoid negative values
arima_forecast_mean = arima_forecast.predicted_mean
arima_forecast_mean[arima_forecast_mean < 0] = 0

sarima_forecast_mean = sarima_forecast.predicted_mean
sarima_forecast_mean[sarima_forecast_mean < 0] = 0

# Plot forecasts
plt.figure(figsize=(12, 6))
plt.plot(sample_data.index, sample_data, label='Observed')
plt.plot(arima_forecast.predicted_mean.index, arima_forecast.predicted_mean, label='ARIMA Forecast')
plt.plot(sarima_forecast.predicted_mean.index, sarima_forecast.predicted_mean, label='SARIMA Forecast')
plt.fill_between(arima_forecast.conf_int().index, arima_forecast.conf_int().iloc[:, 0], arima_forecast.conf_int().iloc[:, 1], color='gray', alpha=0.2)
plt.fill_between(sarima_forecast.conf_int().index, sarima_forecast.conf_int().iloc[:, 0], sarima_forecast.conf_int().iloc[:, 1], color='gray', alpha=0.2)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title(f'Sales Forecast for {sample_sku}')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.forecasting.theta import ThetaModel
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# # Load the data
# file_path = '/mnt/data/100_SKUS.csv'
# data = pd.read_csv(file_path)
# data_long = pd.melt(data, id_vars=['material'], var_name='date', value_name='sales')
# data_long['date'] = pd.to_datetime(data_long['date'], format='%m/%d/%Y')
# data_long = data_long.sort_values(by=['material', 'date'])

# # Select a sample SKU for analysis
# sample_sku = 'sku1009'
# sample_data = data_long[data_long['material'] == sample_sku].set_index('date')['sales']

# Train-test split
train_size = int(len(sample_data) * 0.9)
train, test = sample_data[:train_size], sample_data[train_size:]

# Fit ARIMA model
arima_model = ARIMA(train, order=(2, 1, 2))  # Adjust order as needed
arima_fit = arima_model.fit()

# Fit SARIMA model
sarima_model = SARIMAX(train, order=(2, 1, 2), seasonal_order=(1, 1, 1, 12))  # Adjust order as needed
sarima_fit = sarima_model.fit()

# Fit Exponential Smoothing model (ETS)
ets_model = ExponentialSmoothing(train, seasonal='add', seasonal_periods=12).fit()

# Fit Theta model
theta_model = ThetaModel(train).fit()

# Prepare data for machine learning models
def create_features(df, label=None):
    df['month'] = df.index.month
    df['year'] = df.index.year
    X = df[['month', 'year']]
    if label:
        y = df[label]
        return X, y
    return X

X_train, y_train = create_features(train.to_frame(), label='sales')
X_test, y_test = create_features(test.to_frame(), label='sales')

# Fit XGBoost model
xgb_model = XGBRegressor(objective='reg:squarederror')
xgb_model.fit(X_train, y_train)

# Fit CatBoost model
cat_model = CatBoostRegressor(silent=True)
cat_model.fit(X_train, y_train)

# Fit Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Forecasting
forecast_steps = len(test)
arima_forecast = arima_fit.forecast(steps=forecast_steps)
sarima_forecast = sarima_fit.get_forecast(steps=forecast_steps).predicted_mean
ets_forecast = ets_model.forecast(steps=forecast_steps)
theta_forecast = theta_model.forecast(steps=forecast_steps)

# Predict using machine learning models
xgb_forecast = xgb_model.predict(X_test)
cat_forecast = cat_model.predict(X_test)
lr_forecast = lr_model.predict(X_test)

# Post-process forecasts to avoid negative values
arima_forecast[arima_forecast < 0] = 0
sarima_forecast[sarima_forecast < 0] = 0
ets_forecast[ets_forecast < 0] = 0
theta_forecast[theta_forecast < 0] = 0
xgb_forecast[xgb_forecast < 0] = 0
cat_forecast[cat_forecast < 0] = 0
lr_forecast[lr_forecast < 0] = 0

# Plot forecasts
plt.figure(figsize=(15, 10))
plt.plot(sample_data.index, sample_data, label='Observed')
plt.plot(test.index, arima_forecast, label='ARIMA Forecast')
plt.plot(test.index, sarima_forecast, label='SARIMA Forecast')
plt.plot(test.index, ets_forecast, label='ETS Forecast')
plt.plot(test.index, theta_forecast, label='Theta Forecast')
plt.plot(test.index, xgb_forecast, label='XGBoost Forecast')
plt.plot(test.index, cat_forecast, label='CatBoost Forecast')
plt.plot(test.index, lr_forecast, label='Linear Regression Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title(f'Sales Forecast for {sample_sku}')
plt.legend()
plt.show()

# Evaluation
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluate_forecasts(actual, predicted):
    mse = mean_squared_error(actual, predicted)
    mae = mean_absolute_error(actual, predicted)
    mape = mean_absolute_percentage_error(actual, predicted)
    accuracy = 100 - mape
    return actual, predicted, mse, mae, mape, accuracy

metrics = {}
metrics['ARIMA'] = evaluate_forecasts(test, arima_forecast)
metrics['SARIMA'] = evaluate_forecasts(test, sarima_forecast)
metrics['ETS'] = evaluate_forecasts(test, ets_forecast)
metrics['Theta'] = evaluate_forecasts(test, theta_forecast)
metrics['XGBoost'] = evaluate_forecasts(test, xgb_forecast)
metrics['CatBoost'] = evaluate_forecasts(test, cat_forecast)
metrics['Linear Regression'] = evaluate_forecasts(test, lr_forecast)

metrics_df = pd.DataFrame(metrics, index=['Actual','Predicted','MSE', 'MAE', 'MAPE', 'Accuracy'])
metrics_df
```
